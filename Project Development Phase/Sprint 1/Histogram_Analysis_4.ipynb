```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
```

```python
A = imread('/content/5ee4cb1c-d933-4e89-b364-9b740b48e077___RS_HL 6224.JPG')
B = imread('/content/6a5ded2c-8123-4d42-86fe-7be3c9f268a1___RS_HL 6302.JPG')
```

```python
plt.figure()
plt.subplot(121), plt.imshow(A)
plt.subplot(122), plt.imshow(B)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(np.abs(A[:, :, 0].astype(float) - B[:, :, 0].astype(float)), cmap='gray')
plt.show()
```

```python
d = imread('/content/5ee4cb1c-d933-4e89-b364-9b740b48e077___RS_HL 6224.JPG')
mask = imread('/content/6a5ded2c-8123-4d42-86fe-7be3c9f268a1___RS_HL 6302.JPG')
print(np.amin(d), np.amax(d))
print(np.amin(mask), np.amax(mask))
```
